```python
import pandas as pd
import os
```

```python
DATA_PATH = '../data/'
```

```python
os.path.exists(os.path.join(DATA_PATH, 'data.csv'))
```

```python
df = pd.read_csv(os.path.join(DATA_PATH, 'data.csv'), sep=',')
```

```python
df.head()
```

```python
#get only introduction and understan
introduction_df = df[['NAME', 'CONTINENT','INTRODUCTION', 'UNDERSTAND']]
introduction_df.head()
```

```python
columns_to_merge = ['INTRODUCTION', 'UNDERSTAND']
introduction_df.loc[:, 'INFORMATION'] = introduction_df[columns_to_merge].apply(lambda row: '. '.join(row.dropna()), axis=1)
introduction_df = introduction_df.iloc[1:]
introduction_df = introduction_df.drop(columns=columns_to_merge)
introduction_df.replace('', pd.NA, inplace=True)
introduction_df = introduction_df.dropna().reset_index(drop=True)
introduction_df
```
