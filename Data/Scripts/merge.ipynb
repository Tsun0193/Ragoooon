```python
import pandas as pd
import os
```

```python
DATA_PATH = '../data/'
```

```python
os.path.exists(os.path.join(DATA_PATH, 'data.csv'))
```

```python
df = pd.read_csv(os.path.join(DATA_PATH, 'data.csv'), sep=',')
```

```python
df.head()
```

```python
columns_to_merge = ['UNDERSTAND', 'GET_IN', 'SEE', 'DO', 'BUY', 'EAT', 'DRINK', 'SLEEP', 'STAY_SAFE', 'GO_NEXT']
df['Merged'] = df[columns_to_merge].apply(lambda row: '. '.join(row.dropna()), axis=1)
```

```python
df.head()
```

```python
df = df.iloc[1:].reset_index(drop=True)
df = df.reset_index(drop=True)
df.head()
```

```python
#drop UNDERSTAND, GET_IN, SEE, DO, BUY, EAT, DRINK, SLEEP, STAY_SAFE, GO_NEXT columns
df = df.drop(columns=columns_to_merge)
df.head()
```

```python
df.to_csv('final.csv', index=False)
```
